```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

file = "C:\\Users\\luis3\\OneDrive\\Escritorio\\COLLEGE\\NJIT\\FALL 2020\\CS301 - Intro to Data Science\\Project\\out.csv"
f1 = pd.read_csv(file)
f1.head(100)
```

```python
f1['BOROUGH'].unique()
```

```python
grouped = f1.groupby('BOROUGH')
```

```python
each_borough_injured = grouped[['NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']].agg(np.sum)
each_borough_injured
```

```python
each_borough_killed = grouped[['NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']].agg(np.sum)
each_borough_killed
```

```python

sns.heatmap(each_borough_injured, linewidths=0, annot=False)
plt.title('Headmap of Injured People')
```

```python
sns.heatmap(each_borough_killed, linewidths=0, annot=False)
plt.title('Headmap of Killed People')
```
